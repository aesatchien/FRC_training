### Exercises week 3 - file io and data manipulation
* open the file 'files\co2-mm-mlo_csv.csv' and read all of the lines

```python
with open('files\co2-mm-mlo_csv.csv', 'r') as file:
    csv_data = file.readlines()
csv_data[0:3]
```

```python
# strip the newlines
csv_data = [line.strip().split(',') for line in csv_data]
csv_data[0:3]
```

```python
# remove the header and store
# only do it once or else you need to reread the file
columns = csv_data.pop(0)
columns
```

```python
# make a dictionary of the data
data_records = {}
for ix, line in enumerate(csv_data):
    line_data = csv_data[ix]
    temp_dict = {columns[i]:line_data[i] for i in range(len(columns))}
    data_records.update({ix:temp_dict})
data_records[0]
```

```python
# make a better dictionary of the data, with each column as a list
data_columns = {}
for ix, col in enumerate(columns):
    line_data = [line[ix] for line in csv_data]
    data_columns.update({col:line_data})
```

```python
import pandas as pd
```

```python
pd.DataFrame(data_columns)
```

```python
pd.DataFrame(data_records)
```
